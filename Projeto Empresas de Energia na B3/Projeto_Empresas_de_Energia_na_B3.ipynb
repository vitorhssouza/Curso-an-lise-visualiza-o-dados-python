```python
# Importando bibliotecas
# (importing libraries)

# bibliotecas para modelagem e matrizes
# (libraries for modeling and matrices)
import numpy as np
import pandas as pd

# bibliotecas para análises gráficas
# (libraries for graphical analysis)
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
# Lendo a base de
# (reading the data)
base_dados = pd.read_excel('Dados_Empresas_Energia.xlsx')

# Imprimindo os primeiros registros 
# (firts records)
base_dados.head()
```

```python
# Verificando informações sobre a base de dados
# (check information)
base_dados.info()
```

```python
# Descrição base de dados 
# Description database

base_dados.describe()
```

```python
# Visualizando os campos nulos no gráfico
# OBS: No gráfico abaixo podemos ver que existem campos nulos  
# na coluna 'Year' e 'Genre'
# (Visualizing the null fields in the chart
# OBS: In the chart below we can see that there are null fields
# in the 'Year' and 'Genre' column)


nulo = base_dados.isnull()

plt.figure(figsize=(16, 5))
plt.title('Analise campos nulos')
sns.heatmap(nulo, cbar=False);
```

```python
# Grafico 
# (Graphic)

# Estilo do gráfico
plt.style.use('seaborn-darkgrid')

# Tamanho
plt.figure(figsize=(20, 10))

# titulo
plt.title('Análise ações de empresas de Energia', loc='left', fontsize= 18, fontweight=600);

# Plot Petrobras
plt.plot(base_dados.index, base_dados['Petrobras'], color='#008c4a', linewidth=3, alpha=0.7)

# Texto Petrobras
plt.text(base_dados.index[-1], base_dados['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='left')

# plot de todas as colunas 
for colunas in base_dados.columns[1:]:
    # Plot das ações
    plt.plot(base_dados.index, base_dados[colunas], color='gray', linewidth=1.5, alpha=0.7)
    # Texto das outras ações
    plt.text(base_dados.index[-1], base_dados[colunas].tail(1), colunas, color='gray', size='large', horizontalalignment='left')
  

# Labels 
plt.xlabel('Período', fontsize= 14)
plt.ylabel('Preço de Fechamento (R$)', fontsize= 14);
```
