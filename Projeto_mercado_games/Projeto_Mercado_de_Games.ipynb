## **Conjunto de dados de vendas de videogames**

Nesse projeto iremos utilizar python e sua biblioteca para análise e visualização de dados com intuito de analisar e visualizar as vendas de 
jogos de playstation4 pelo o mundo. 

**Sobre o conjunto de dados**
Dados de vendas e classificações de jogos de videogame extraídos do VzCharts

[Download dos Dados](https://www.kaggle.com/spscientist/students-performance-in-exams)

```python
# Importando bibliotecas
# (importing libraries)

# bibliotecas para modelagem e matrizes
# (libraries for modeling and matrices)
import numpy as np
import pandas as pd

# bibliotecas para análises gráficas
# (libraries for graphical analysis)
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar e desabilitar avisos
# (Lib to ignore and disable warnings)
import warnings
warnings.filterwarnings('ignore')
```

```python
# Lendo a base de
# (reading the data)
base_dados = pd.read_csv('PS4_GamesSales.csv', encoding='latin-1')

# Imprimindo os primeiros registros 
# (firts records)
base_dados.head()
```

```python
# Verificando informações sobre a base de dados
# (check information)
base_dados.info()
```

```python
# Visualizando os campos nulos no gráfico
# OBS: No gráfico abaixo podemos ver que existem campos nulos  
# na coluna 'Year' e 'Genre'
# (Visualizing the null fields in the chart
# OBS: In the chart below we can see that there are null fields
# in the 'Year' and 'Genre' column)


nulo = base_dados.isnull()

plt.figure(figsize=(16, 5))
plt.title('Analise campos nulos')
sns.heatmap(nulo, cbar=False);
```

```python
# Retirando os valores nulos com o inplace ele ja troca diretamente na base de dados
# (Removing the null values ​​with inplace it already changes directly in the database)
base_dados.dropna(inplace=True)
```

```python
# Descrição base de dados 
# Description database

base_dados.describe()
```

```python
# Analise de quanto foi o valor de vendas no ano
# (Analyze how much was the sales value in the year)

# Tamanho da imagem
# (Image size)
plt.figure(figsize=(10,5))

# Titulo
# (Title)
plt.title('Quantidade de Vendas Globais (mi)', loc='left', fontsize=14)

# Grafico
# (Graphic)
sns.barplot( data=base_dados, x='Year', y='Global', ci=None, color='#69b3a2', estimator=sum)

# Label
plt.ylabel('Quantidade Vendas');
```

```python
# Retirando os anos vazio que são 2019 e 2020
# (Removing the empty years which are 2019 and 2020)

base_dados = base_dados.loc[(base_dados['Year'] != 2019) & (base_dados['Year'] != 2020)]
base_dados.head()
```

```python
# Distribuição do valor global 
# (Distribution of the global value)

# Tamanho
# (size)
plt.figure(figsize=(12,5))

# Estilo 
# (Style)
plt.style.use('ggplot')

# Titulo
# (Title)
plt.title('Distribuição das Vendas Globais', loc='left', fontsize=16)

# Plot
sns.kdeplot(base_dados['Global'], shade=True, bw=1, color='#96a8a8', linewidth=2.5);
```

```python
# Utilizando o groupby para agrupar por ano 
# (Using groupby to group by year)
analise = base_dados.groupby(by=['Year']).sum().reset_index()
analise
```

```python
# Grafico boxplot para analise de vendas no ano
# (Boxplot chart to analyze sales in the year)

# Tamanho
# (Size)
plt.figure(figsize=(12,5))

# Titulo
# (Title)
plt.title('Análise da Distribuição Global (mi)')

# Plot
sns.boxplot(data=base_dados, x='Year', y='Global'); 
```

```python
# Analisando a proporção dos 100% de cada continente comparado ao total
# (Analyzing the proportion of the 100% of each continent compared to the total)

america = [(america / total) * 100 for america, total in zip(analise['North America'], analise['Global'])]

europa = [(europa / total) * 100 for europa, total in zip(analise['Europe'], analise['Global'])]

japao = [(japao / total) * 100 for japao, total in zip(analise['Japan'], analise['Global'])]

mundo = [(mundo / total) * 100 for mundo, total in zip(analise['Rest of World'], analise['Global'])]

america, europa, japao, mundo
```

```python
# Tamanho
# (Size)
plt.figure(figsize=(12,5))

# largura barra no gráfico
# (bar width on chart)
largura_barra = 0.85

grupos = [0, 1, 2, 3, 4, 5]

# Titulo
# (Title)
plt.title('Análise da Distribuição Por Continentes')

# plot america
plt.bar(grupos, america, width=largura_barra, color='#b5ffb0', edgecolor='white')

# plot da europa
plt.bar(grupos, europa, bottom=america, width=largura_barra, color='#f9bc86', edgecolor='white')

# plot japão
plt.bar(grupos, japao, bottom=[a + b for a, b in zip(america, europa)], width=largura_barra, color='#a3acff', edgecolor='white')

# plot resto do mundo
plt.bar(grupos, mundo, bottom=[a + b + c for a, b, c  in zip(america, europa, japao)], width=largura_barra, color='#d3acfe', edgecolor='white');

plt.xticks(grupos, analise['Year']);
plt.xlabel('Grupo');
plt.xlabel('Distribuição %');

# Legenda
# (Legend)
plt.legend(['America N', 'Europa', 'Japão', 'Mundo'], loc='uper left', bbox_to_anchor=(0.75, -0.1), ncol=4);
```

```python
# Empresas que publicaram jogos para o ps4
# (Companies that published games for the ps4)

base_dados['Publisher'].unique()
```

```python
# Convertendo os dados acima para melhor plotagem do grafico. Vou transforma os valores em texto em numericos 
# (Converting the above data for better graph plotting. I will transform the values ​​in text into numerics)

from sklearn.preprocessing import LabelEncoder

funcao_label = LabelEncoder()

base_dados['Produto'] = funcao_label.fit_transform(base_dados['Publisher'])
base_dados['Genero'] = funcao_label.fit_transform(base_dados['Genre'])
base_dados['Jogo'] = funcao_label.fit_transform(base_dados['Game'])

base_dados.head()
```

```python
# Gerando paletas de cores
# (Generating color palettes)
paleta_cores = sns.color_palette('Dark2', 8)
paleta_cores
```

```python
# Analise dos produtores de game 
# (Analysis of game developers)

plt.figure(figsize=(20, 10))
plt.title('Análise dos Produtores de Games (mi)')

sns.scatterplot(data = base_dados, x='Produto', y='Global', color=paleta_cores[0]);
```

```python
# Analise dos generos de game 
# (Analysis of game genres)

plt.figure(figsize=(20, 10))
plt.title('Análise dos Generos de Games (mi)')

sns.scatterplot(data = base_dados, x='Genero', y='Global', color=paleta_cores[1]);
```

```python
# Análise dos games
# (Game analysis)

plt.figure(figsize=(20, 10))
plt.title('Análise dos Games (mi)')

sns.scatterplot(data = base_dados, x='Jogo', y='Global', color=paleta_cores[3]);
```

```python
# Relatorio contendo os gráfico utilizado para visualização dos dados acima
# (Report containing the graph used to visualize the above data)

# Tamanho da imagem
fig, ax = plt.subplots(figsize=(18, 15))
colunas = 2
linhas = 3

# cor de fundo
cor_fundo = '#f5f5f5'
ax.set_facecolor(cor_fundo)
fig.set_facecolor(cor_fundo)

# Estilo dos graficos 
plt.style.use('seaborn')

# incluindo subtitulo
plt.suptitle('Python para Análise de Dados \n Projeto prático 5 - Análise Mercado de Games PS4', fontsize=22, color='#404040', fontweight=600)

# Parametros para os grid 

for c in range(1, 7):
  plt.subplot(linhas, colunas, c)
  if c == 1:
      # Titulo
      plt.title('Quantidade de Vendas Globais (mi)', loc='left', fontsize=14)

      # Grafico
      plt.bar( base_dados['Year'], base_dados['Global'], color='#69b3a2')

      # Label
      plt.ylabel('Quantidade Vendas');

  elif c == 2:
      # Acessando gráfico 2
      # Titulo
      plt.title('Análise da Distribuição Global (mi)')

      # Plot
      sns.boxplot(data=base_dados, x='Year', y='Global');

  elif c == 3:
      # largura barra no gráfico
      largura_barra = 0.85
      grupos = [0, 1, 2, 3, 4, 5]
      # Titulo
      plt.title('Análise da Distribuição Por Continentes', loc='left', fontsize=14)
      # plot america
      plt.bar(grupos, america, width=largura_barra, color='#b5ffb0', edgecolor='white')
      # plot da europa
      plt.bar(grupos, europa, bottom=america, width=largura_barra, color='#f9bc86', edgecolor='white')
      # plot japão
      plt.bar(grupos, europa, bottom=[a + b for a, b in zip(america, europa)], width=largura_barra, color='#a3acff', edgecolor='white')
      # plot resto do mundo
      plt.bar(grupos, europa, bottom=[a + b + c for a, b, c  in zip(america, europa, japao)], width=largura_barra, color='#d3acfe', edgecolor='white');
      plt.xticks(grupos, analise['Year']);
      plt.xlabel('Grupo');
      plt.xlabel('Distribuição %');
      # legenda
      plt.legend(['America N', 'Europa', 'Japão', 'Mundo'], loc='uper left', bbox_to_anchor=(0.75, -0.1), ncol=4);

  elif c == 4:
      plt.title('Análise dos Produtores de Games (mi)', loc='left', fontsize=14)
      sns.scatterplot(data = base_dados, x='Produto', y='Global', color=paleta_cores[0]);

  elif c == 5:
      plt.title('Análise dos Generos de Games (mi)', loc='left', fontsize=14)
      sns.scatterplot(data = base_dados, x='Genero', y='Global', color=paleta_cores[1]);
  else:
      plt.title('Análise dos Games (mi)', loc='left', fontsize=14)
      sns.scatterplot(data = base_dados, x='Jogo', y='Global', color=paleta_cores[3]);


# Ajustando o layout
plt.subplots_adjust(hspace=0.35, wspace=0.15 )

# Rodapé
rodape = '''
  Esse relatório foi elaborado no treinamento "Python para Análise de Dados"
  está dispónivel no canal do youtube @Data Viking
  by: Vitor Souza
'''

# Incluindo o rodapé no relatório
fig.text(0.5, -0.01, rodape, ha='center', va='bottom', size=12, color='#938ca1');
```
