## **Conjunto de dados salariais - Regressão linear simples**
Nesse projeto iremos utilizar python e sua biblioteca para análise e visualização de dados com intuito de analisar e visualizar se o tempo de experiencia vs salário

[Download dos Dados](https://www.kaggle.com/datasets/karthickveerakumar/salary-data-simple-linear-regression?datasetId=10624&sortBy=voteCount)

```python
# Importando bibliotecas
# (importing libraries)

# bibliotecas para modelagem e matrizes
# (libraries for modeling and matrices)
import numpy as np
import pandas as pd

# bibliotecas para análises gráficas
# (libraries for graphical analysis)
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar e desabilitar avisos
# (Lib to ignore and disable warnings)
import warnings
warnings.filterwarnings('ignore')
```

```python
# Lendo a base de
# (reading the data)
base_dados = pd.read_csv('Salary_Data.csv')

# Imprimindo os primeiros registros 
# (firts records)
base_dados.head()
```

```python
# Verificando informações sobre a base de dados
# (check information)
base_dados.info()
```

```python
# Visualizando os campos nulos no gráfico
# OBS: No gráfico abaixo podemos ver que existem campos nulos  
# na coluna 'Year' e 'Genre'
# (Visualizing the null fields in the chart
# OBS: In the chart below we can see that there are null fields
# in the 'Year' and 'Genre' column)


nulo = base_dados.isnull()

plt.figure(figsize=(16, 5))
plt.title('Analise campos nulos')
sns.heatmap(nulo, cbar=False);
```

```python
# Análise da renda 
# (Income analysis)
plt.figure(figsize=(10,6))
plt.title('Análise da Renda')
sns.kdeplot(base_dados['Salary'], color='green');
```

```python
plt.figure(figsize=(10,6))
plt.title('Análise de Xp')
sns.kdeplot(base_dados['YearsExperience'], color='blue');
```

```python
# Distribuição
# (Distribution)

plt.figure(figsize=(10,6))
sns.boxplot(base_dados['Salary']);
```

```python
# Distribuição
# (Distribution)
plt.figure(figsize=(10,6))
sns.boxplot(base_dados['YearsExperience']);
```

```python
# Distribuição
# (Distribution)

plt.figure(figsize=(10,6))
sns.scatterplot(data=base_dados, x='Salary', y='YearsExperience');
```

```python
# Distribuição
# (Distribution)
sns.regplot(data=base_dados, x='Salary', y= 'YearsExperience');
```

```python
# correlação
# (account)

correlacao = base_dados.corr('pearson')
sns.heatmap(correlacao, annot=True, cbar=False);
```

```python
# correlação
# (account)

correlacao = base_dados.corr('spearman')
sns.heatmap(correlacao, annot=True, cbar=False);
```
